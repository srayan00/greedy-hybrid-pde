```python
import numerical_solver
import numpy as np
import pde
```

```python
def f(x):
    return np.sin(np.pi * x)
def a(x):
    return 1
x = np.linspace(0, 1, 11)
equation = pde.PoissonEquation1D(a, f, "Dirichlet", x)
solver = numerical_solver.MultigridSolver(equation)
# print(equation.A)
# print(f"The true solution is {f(x)/(np.pi**2)}")
# # print(solver.restrictor)
# print(solver.interpolator)
solver.iteration(np.zeros_like(equation.b))
```

```python
def f(x, y):
    return np.sin(np.pi * x) * np.sin(np.pi * y)
def a(x, y):
    return 1
x = np.linspace(0, 1, 7)
y = np.linspace(0, 1, 7)
equation2d = pde.PoissonEquation2D(a, f, "Dirichlet", x, y)
print(equation2d.A[9].reshape((7, 7)))
u_exact_2d = (1/(2*np.pi**2)) * np.sin(np.pi * x[:, None]) * np.sin(np.pi * y[None, :])
print(f"The true 2D solution is {f(x[:, None], y[None, :])/(2*np.pi**2)}")
print(f"The numeric 2D solution is {equation2d.solve()}")
print(f"error norm is {np.linalg.norm(equation2d.solve() - f(x[:, None], y[None, :])/(2*np.pi**2))}")
```
